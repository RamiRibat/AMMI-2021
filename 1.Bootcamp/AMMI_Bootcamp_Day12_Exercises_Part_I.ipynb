Exercise 2

```
t = 1; v0 = 1; g =10
```

```
# Position function f(t)
f = lambda t, v0, g: v0*t - 0.5*g*t**2
f(t, v0, g)
```

```
def differentiation(f, t, v0, g, h=0.01, order=1):
  diff = lambda t, v0, g: (f(t+h, v0, g) - f(t-h, v0, g))/(2*h)
  if order == 1:
    return diff
  else:
    return lambda t, v0, g: (diff(t+h, v0, g) - diff(t-h, v0, g))/(2*h)
```

```
# Velocity function f(t)
df = differentiation(f, t, v0, g, order=1)
df(t, v0, g)
```

```
# Acceleration function f(t)
ddf = differentiation(f, t, v0, g, order=2)
ddf(t, v0, g)
```

```
def kinematics(t, v0, g):
  x = f(t, v0, g)
  v = df(t, v0, g)
  a = ddf(t, v0, g)
  return x, v, a
```

```
T = [0.05*i for i in range(10)]
v0 = 5; g = 9.81
X = []; V = []; A = []
for t in T:
  x, v, a = kinematics(t, v0, g)
  X.append(x)
  V.append(v)
  A.append(a)

plt.figure(figsize=(6,4), dpi=100)
bar_width = 0.01
plt.bar([t - bar_width for t in T], X, width=bar_width, label='Position')
plt.bar(T, V, width=bar_width, label='Velocity')
plt.bar([t + bar_width for t in T], A, width=bar_width, label='Acceleration')
plt.xticks(T)
plt.yticks(range(-10,10))
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

```
def threeSum(L, n):
  L.sort()
  P = [] # list of triples
  for k in range(len(L)-2):
    if k>0 and L[k] == L[k-1]:
      continue
    start = k+1
    end = len(L)-1

    while start < end:
      if (L[k] + L[start] + L[end]) < n:
        start += 1
      elif (L[k] + L[start] + L[end]) > n:
        end -= 1
      else:# (L[k] + L[start] + L[end]) == n:
        P.append([L[k], L[start], L[end]])
        while start < len(L)-1 and L[start] == L[start+1]: start += 1
        while end > 0 and L[end-1] == L[end]: end -= 1
        start += 1
        end -= 1

  return P
```

```
threeSum(L1, 0)
```

```
threeSum(L2, 0)
```
