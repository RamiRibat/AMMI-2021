### Exercise 1:  Bisection Method

```python
# 1.1
def Bisection1(fun, start, end, error):
    while abs(start - end) > error:
        mid = (start + end)/2
        if fun(start)*fun(mid) < 0: end = mid
        elif fun(start)*fun(mid) > 0: start = mid
        else: break
        plt.plot(mid, fun(mid), '.m', markersize=7.5)
    root = mid
    plt.axvline(mid, color='green')
    plt.plot(mid, fun(mid), '*r', markersize=10)
    return root
```

```python
def linspace(start, end, n):
    if n == 1:
        yield end
        return
    h = (end - start) / (n - 1)
    for i in range(n):
        yield start + h * i
```

```python
# 2.1
# Function
fun1 = lambda x: x**2 - 1
```

```python
# 2.2
# Root 1
start = -10
end = 0
n = 100
error = 1e-6
X = list(linspace(start, end, n))
Y = [fun1(x) for x in X]
plt.plot(X, Y)
# plt.plot(0, 0, '*g')
plt.axhline(0, color='black')

root = Bisection1(fun1, start, end, error)
print(f'Root 1 of f(x) = {root}')
```

```python
# 2.3
# root 2
start = 0
end = 10
n = 100
error = 1e-6
X = list(linspace(start, end, n))
Y = [fun1(x) for x in X]
plt.plot(X, Y)
plt.axhline(0, color='black')

root = Bisection1(fun1, start, end, error)
print(f'Root 2 of f(x) = {root}')
```

### Exercise 2:  Sphere

```python
# Radius
r = 5
```

```python
# 1. Volume = (4/3)*pi*r^3
```

```python
vol = lambda r: (4/3)*pi*r**3
vol(5)
```

```python
# 2. Derivative
```

```python
h = 0.01
derivative = lambda f, x: (f(x + h) - f(x - h))/(2*h)
derivative(vol, 5)
```

```python
# 3. Surface Area = 4*pi*r^2
```

```python
area = lambda r: 4*pi*r**2
area(5) # area = derivative of volume
```

```python
# 4. vol + Area
```

```python
def myArea(r):
    V = vol(r)
    A = derivative(vol, r)
    print(f'For radius {r}, the sphere volume = {V}, and area = {A}')
#     return V, A
```
