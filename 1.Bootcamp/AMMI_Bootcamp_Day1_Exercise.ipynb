```python
#04
def fun4(p1, p2):
  del_x = int(p2[1]) - int(p1[1])
  del_y = int(p2[3]) - int(p1[3])
  distance = ((del_x**2)+(del_y**2))**(1/2)
  return distance
fun4(p1, p2)
```

```python
#05
def fun5(x, y):
  if x % 2 != 0 and y % 2 != 0:
    print(f'odd numbers,  Product  --> {x}*{y}={x*y}')
  elif x % 2 ==0 and y % 2 ==0:
    print(f'Even numbers, Sum     --> {x}+{y}={x+y}')
  else:
    print(f'Odd \& Even,  Div      --> {x}/{y}={x/y}')
fun5(3, 9)
fun5(2, 6)
fun5(5, 8)
```
