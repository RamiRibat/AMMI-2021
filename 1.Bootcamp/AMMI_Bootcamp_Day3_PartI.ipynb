### Exercise 3:  Normal

```python
m = 0
s = 1
n = 50

a = m-5*s
b = m+5*s
h = (b-a)/(n-1)
X = [(a + i*h) for i in range(n)]


def normal(x, m=0, s=1):
    return ((2*(pi**0.5)*s)**-1) * exp(-0.5*((x-m)/s)**2)

for x in X:
    print(f'Normal({x}) = {normal(x)}')
```

```python
import matplotlib.pyplot as plt
Y = [normal(x) for x in X]
plt.plot(X, Y, 'k', linewidth=2)
```

### Exercise 4:  Factorial

```python
def fact(n):
    factorial = 1
    for i in range(1,n+1): 
        factorial = factorial * i
    return factorial
print(fact(0))
print(fact(3))
print(fact(6))
```

### Exercise 5:  MinMax

```python
n = 100
a = -5; b = 5
h = (b-a)/(n-1)
X = [(a + i*h) for i in range(n)]
Y = [sin(x) for x in X]
```

```python
# %%timeit
def MinMax(Y):
    minimum = Y[0]
    maximum = Y[0]
    for i in Y[1:]:
        if i < minimum: minimum = i
    for i in Y[1:]:
        if i > maximum: maximum = i
    return minimum, maximum
minimum, maximum = MinMax(Y)
print(f'Minimum is {minimum}, and Maximum is {maximum}')
```

### Exercise 7:  Complementary

```python
def comp(dna):
    for i in range(len(dna)):
        if dna[i] == 'A': rev = 'T'
        elif dna[i] == 'T': rev = 'A'
        elif dna[i] == 'G': rev = 'C'
        else: rev = 'G'
        print(rev, end='')
comp('ACGTCAAAAAAA')
```
