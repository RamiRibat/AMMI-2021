## Exercise 2: Pairwise sum
<b> 
Given two lists l1 and l2 of size 6, create a new list that contain the pairwise sum. 

Example: If we have l1 = [2, 3, 4] and l2 = [4, 5, 7], it should print l3 = [6, 8, 11]

</b> 


```python
##### WRITE YOUR CODE HERE #####
l1 = [2, 3, 4]
l2 = [4, 5, 7]
l3 = []
for a,b in zip(l1, l2):
  l3.append(a + b)
print(l3)
#### END CODE ####
```

## Exercise 3: Intersection of two lists
<b> 
Given two lists l1 and l2 of , print the intersection of the two lists. Lists don't contain duplicates.

Example: If we have l1 = [1, 3, 2, 4] and l2 = [3, 4, 5, 7], it should print : 3,4

</b> 


```python
# l1 = [1,2,3,4,5,6]; l2 = [2,3,9,16,1]
l1 = [1, 3, 2, 4]; l2 = [3, 4, 5, 7]
##### WRITE YOUR CODE HERE #####
inx = []
for i, a in enumerate(l1):
  for j, b in enumerate(l2):
    if a == b:
      inx.append(i)
ints = [l1[k] for k in inx]
print(f'Intersections: {ints}')  
#### END CODE ####
```

## Exercise 4: Palindrome
<b> Given a string s, check whether s is a palindrome. s is the palindrome if s is equal to its reverse. Ps: Use a loop, slicing operation is not allowed.</b>


```python
s = 'abababa'
palind = False
##### WRITE YOUR CODE HERE #####
for i in range(0, len(s)):
  if s[i] != s[-1-i]:
    palind = False
    break
  else:
    palind = True
if palind:
  print('String s is palindrome')
else:
  print('String s is NOT palindrome')
#### END CODE ####
```

## Exercise 7: Prime Number
<b> Given an integer n, check whether n is a prime number. A number is a prime number if it is divisible only by 1 and itself.</b> </br>


```python
n = 23
prime = True
##### WRITE YOUR CODE HERE #####
for i in range(2, n):
  if n % i == 0:
    prime = False
    
if prime:
  print(f'{n} is a prime number')
else:
  print(f'{n} is not a prime number')
#### END CODE ####
```

## Exercise 8: Two sum
<b> 
Given an integer n and a sorted list l, find two elements that sum up to n and their indexes. We can have many combinations but just output one of them.

Example: l = [1,2,3,4], n = 5 ==> combination = 1,4
</b> 



```python
# %%timeit
l = [1,2,3,4,5]
n = 9
##### WRITE YOUR CODE HERE #####
# ----------Version 01----------
for i in l:
  for j in l:
    if i != j and (i + j) == n:
      break
print(f'Combination = {i, j}')
```

```python
# %%timeit
l = [1,2,6,3,4,5]
n = 9
# ----------Version 02----------
start = 0
end = len(l)-1
while start < end:
  if l[start] + l[end] == n:
    print(f'Bingo! combination is ({l[start]}, {l[end]})')
    break
  elif l[start] + l[end] > n:
    end -= 1
  else:
    start += 1
#### END CODE ####
```

## Exercise 9: Power of Two
<b> Given an integer n, check whether n is a power of two. </b>


```python
##### WRITE YOUR CODE HERE #####
n = 64

# ----------Version 01----------
m = n
power = 1
while True:
  if m / 2 == 2 and m % 2 == 0:
    power += 1
    print(f'{n} is a power of two: 2^{power}')
    break
  elif m % 2 == 0:
    m /= 2
    power += 1
  else:
    print(f'{n} is NOT a power of two!')
    break

# ----------Version 02----------
if n > 0 and (n & (n-1)) == 0:
  print(f'{n} is a power of two!')
else:
  print(f'{n} is NOT a power of two!')

#### END CODE ####
```
