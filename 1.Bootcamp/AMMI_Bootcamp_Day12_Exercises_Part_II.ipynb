## Exercise 1:

You are given a `sorted` array. Remove the duplicates such that each element appears only once and return a new list. You can allocate new memory space.<br>

Ex: `arr = [1, 1, 1, 2, 2, 3] ==> [1, 2, 3]`


```python
def removeDuplicates(arr):
    
    ##### WRITE YOUR CODE HERE ###
    arr_new = []
    for i in range(len(arr)-1):
      if arr[i] != arr[i+1]:
        arr_new.append(arr[i])
    arr_new.append(arr[i+1])
    #### END OF YOUR CODE #####
    
    return arr_new
```

```python
a = [1, 1, 1, 2, 2, 3]
removeDuplicates(a)
```

```python
def removeDuplicates_bis(arr):
    
    ##### WRITE YOUR CODE HERE ###
    i = 0
    while True:
      if i == len(arr)-1:
        break
      if arr[i] == arr[i+1]:
        arr.pop(i+1)
      elif arr[i] != arr[i+1]:
        i += 1
    #### END OF YOUR CODE #####
    
    return arr
```

```python
a = [1, 1, 1, 2, 2, 3]
removeDuplicates_bis(a)
```

## Exercise 3:

You are given a binary matrix. We want to flip horizontally every vector in the matrix and then invert it.  Return the final result.

Ex: `A = [[1,1,0],[1,0,1],[0,0,0]]`

`Step 1` (flip horizontally): `[[0,1,1],[1,0,1],[0,0,0]]` <br><br>
`Step 2` (invert: 0 becomes 1 and 1 becomes 0): `[[1,0,0],[0,1,0],[1,1,1]]`

```python
def flip(matrix):
    for row in matrix:
      for i in range(1, (len(row))):
        row.append(row.pop(-i-1))
    return matrix
```

```python
def inv(matrix):
    for row in matrix:
      for i in range((len(row))):
        if row[i] == 0:
          row[i] = 1
        else:
          row[i] = 0
    return matrix
```

```python
A = [[1,1,0],[1,0,1],[0,0,0]]
```

```python
A_flipped = flip(A)
```

```python
A_flipped_inv = inv(A_flipped)
```

```python
A_flipped_inv
```

## Exercise 4:

You are given a `rotated` array which is rotated `k times`. The array was initially sorted. Write a function that finds the number `k` of rotations.

Ex: `[4, 5, 1, 2, 3]` ==> `k=2`. The initial sorted array is then `[1, 2, 3, 4, 5]`. <br>
First rotation: `[5, 1, 2, 3, 4]`. <br>
Second rotation: `[4, 5, 1, 2, 3]`. <br>

```python
from copy import deepcopy
def find_k(arr):
    #### WRITE YOUR CODE HERE
    arr_ = deepcopy(arr)
    arr_.sort()
    k = 0
    while True:
      arr.append(arr.pop(0))
      k += 1
      if arr == arr_:
        break
    #### END OF YOUR CODE
    return k
```

```python
Ai = [4, 5, 1, 2, 3]
```

```python
find_k(Ai)
```
